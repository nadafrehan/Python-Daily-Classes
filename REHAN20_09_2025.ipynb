```python
# Compile time / Method Overloading Polymorphism
class BCA_1st_Semester :
    def student(self,a,b=0,c=0,d=0):
        return a+b+c+d

BCA = BCA_1st_Semester()
print(BCA.student(5))
print(BCA.student(10,7,8))
print(BCA.student(-10,4,9,-5))
print(BCA.student(99,0,4,-12))
```

```python
class Animal :
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Bow Bow Bow Bow")

class Cat(Animal):
    def sound(self):
        print("Meow Meow Meow")

class Tiger(Animal):
    def sound(self):
        print("Roar Roar Roar")

class Cow(Animal):
    def sound(self):
        print("Ambbbbbaaaaaa")


dog = Dog()
cat = Cat()
tiger = Tiger()
cow = Cow()

dog.sound()
cat.sound()
tiger.sound()
cow.sound()
```

```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def sound(self):
        pass

class Car(Vehicle):
    def sound(self):
        print("I am driving a car")

obj = Car()
obj.sound()
```

```python
#Abstraction examples
from abc import ABC, abstractmethod
class Fruit(ABC):

  #Normal function / method
  def Actiong(self):
    print("I am doing work")

  #abstract Method / function
  @abstractmethod
  def BGMI(self):
    pass # Abstract methods don't have implementation in the base class

  #normal Function / Method
  def Ludo_King(self):
      print("i m win the match")

  #normal function
  def chess(self):
    print("i m district player")

# You need to create a concrete class that inherits from Fruit and implements the abstract method
class MyFruit(Fruit):
    def BGMI(self):
        print("i got royal pass diamond")

    # You need to define the freefire method if you want to call it
    def freefire(self):
        print("I am playing freefire")


# Now you can create an object of the concrete class
obj = MyFruit()
obj.BGMI()
obj.Ludo_King()
obj.chess()
obj.freefire()
```
